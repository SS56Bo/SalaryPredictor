```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtl
from sklearn.linear_model import LinearRegression 
```

```python
dB=pd.read_csv('train.csv')
```

```python
#Preparing Data
X= dB["Experience"].values
y=dB["Salary"].values
```

```python
#Visualize the data

mtl.scatter(X,y)
mtl.grid()
mtl.xlabel("Experience of Employee")
mtl.ylabel("Salary of EMployee")
mtl.show()
```

```python
#reshaping the algorithm to a single feature
X1=np.array(X).reshape(-1,1)
y1=np.array(y).reshape(-1,1)
```

```python
#selecting a linear model
model = LinearRegression()
```

```python
#predictong the datasets into the model
model.fit(X1,y1)
```

```python
#visualizing data
y_new = model.predict(X1)
mtl.scatter(X1,y1)
mtl.plot(X1, y_new, color='red')
mtl.xlabel("Experience of Employee")
mtl.ylabel("Salary of EMployee")
mtl.show()
```

```python
# added an input function that asks the user to enter their experience 
# and predicts their salary
a=float(input("Enter salary: "))
X_new = [[a]]
X_fit=model.predict(X_new)
print("Estimated salary: ", X_fit)
```
